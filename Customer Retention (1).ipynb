```python
data["Group"].value_counts()
```

```python
def group_bar_plot(data,col_name):
    grouped_data = data.groupby(["Group",col_name]).count().reset_index()[["Group",col_name,"user_id"]]
    grouped_data.columns = ["Group",col_name,"Count"]
    grouped_data.loc[(grouped_data["Group"]=="H_Group"),"Normalized_Count"] = grouped_data["Count"]/len(H_df["user_id"].unique())
    grouped_data.loc[(grouped_data["Group"]=="U_Group"),"Normalized_Count"] = grouped_data["Count"]/len(U_df["user_id"].unique())
    grouped_data.loc[(grouped_data["Group"]=="N_Group"),"Normalized_Count"] = grouped_data["Count"]/len(N_df["user_id"].unique())
    grouped_data.set_index(["Group",col_name])[["Normalized_Count"]].plot.bar()
    plt.show()
```
